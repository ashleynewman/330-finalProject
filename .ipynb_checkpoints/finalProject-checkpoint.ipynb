```python
import json
import os
import requests
import re
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from bs4 import BeautifulSoup
import requests
import numpy as np
from scipy.stats import pearsonr
import scipy.stats as stats
```

**Setting Up/Reading in the Data**

```python
def read_streaming_platforms_data():
    #convert csv file to dataframe and return the dataframe
    df = pd.read_csv('MoviesOnStreamingPlatforms_updated.csv')
    return df
```

```python
#calling the function and setting it to a variable stremaing_df
streaming_df = read_streaming_platforms_data()

#switch year column to be string objects
streaming_df['Year'] = streaming_df['Year'].astype(str)

#switch rotten tomatoes column to be string objects
streaming_df['Rotten Tomatoes'] = streaming_df['Rotten Tomatoes'].astype(str)

#swicth imdb column to be float
streaming_df['IMDb'] = streaming_df['IMDb'].astype(float)

streaming_df.tail()
```

```python
def beautiful_soup_imdb():
    #get the website into BeautifulSoup 
    r = requests.get('https://www.imdb.com/chart/moviemeter/?sort=ir,desc&mode=simple&page=1')
    soup = BeautifulSoup(r.text, 'html.parser')
    
    #grabbing the IMDB table
    x = soup.find('table', class_ = 'chart full-width')
    body = x.find('tbody', class_='lister-list')

    #grabbing the necessary rows inside the IMDB table
    titles = body.find_all('td', class_='titleColumn')
    ratings = body.find_all('td', class_='ratingColumn imdbRating')
    year = body.find_all('td', class_='ratingColumn imdbRating')

    #creating lists to pass this data into
    titles_list = []
    ratings_list = []
    year_list = []

    #passing the data from the Beautiful Soup into the lists
    for i in titles:
        temp = i.find('a').text
        titles_list.append(temp)
        year = str(i.find('span', class_='secondaryInfo').text)[1:-1]
        year_list.append(year)
    for i in ratings:
        temp = str(i.find('strong'))
        
        if temp != 'None':
            ratings_list.append(temp[-12:-9])
        else:
            ratings_list.append(np.nan)
    
    #return the lists so they can be read into a dataframe
    return titles_list, ratings_list, year_list
```

```python
def read_imdb_website_data():
    #call the Beautiful Soup function
    titles_list, ratings_list, year_list = beautiful_soup_imdb()

    #read the lists into a dictionary
    data = {'Title': titles_list, 'Year': year_list, 'IMDb': ratings_list}

    #create a dataframe of the lists and return the dataframe
    df = pd.DataFrame(data, columns = ['Title', 'Year', 'IMDb'])
    return df

imdb_df = read_imdb_website_data()
#convert ratings to floats
imdb_df['IMDb'] = imdb_df['IMDb'].astype(float)
imdb_df.tail()
```

**Editing the Data Types**

```python
def convert_years(df):
    #convert dataframe year column to datetime format
    df['Year'] = pd.to_datetime(df['Year'])
    
    return df

#convert years for streaming_df
streaming_df2 = convert_years(streaming_df)
display(streaming_df2.head())

#convert years for imdb_df
imdb_df2 = convert_years(imdb_df)
display(imdb_df2.head())
```

```python
#function to convert percentage string to a float score out of 10
def make_num(x):
    num = x[:-1]
    try:
        return float(num) / 10.0
    except:
        return float(np.nan)

    
# convert the rotten tomatoes column in streaming_df to a score out of 10 instead of a percentage
def convert_rotten_tomatoes_to_score(df):
    df['Rotten Tomatoes'] = df['Rotten Tomatoes'].apply(make_num)
    return df


streaming_df3 = convert_rotten_tomatoes_to_score(streaming_df2)
streaming_df3.head()
```

**Statistical Analysis**

```python
def statistical_analysis(df1, df2):
    #imdb rating series
    rating = df1['IMDb']
    rating = rating.dropna()

    #streaming rotten tomatoes series
    tomato = df2['Rotten Tomatoes']
    tomato = tomato.dropna()
    
    #making lengths the same for series (for each time function is called)
    if len(rating) == 85:
        tomato = tomato[:85]
        
    elif len(tomato) == 5158:
        rating = rating[:5158]
        
    #correlation analysis
    corr, _ = pearsonr(rating, tomato)
    
    print('Pearsons correlation: %.3f' % corr)
    

    
#How are IMDb rating and rotten tomatoes score related?
#
statistical_analysis(imdb_df2, streaming_df3)

#how are IMDB rating and rotten tomatoes score related in regard to same file?
#anticorrelated b/c negative
statistical_analysis(streaming_df3, streaming_df3)
```

```python
def statistical_analysis2(x, y):
    #tells us whether 2 data samples have different means
    x = x.dropna()
    y = y.dropna()
    
    print(stats.ttest_ind(a=x, b=y, equal_var=False))
    
    
#t-test for checking for IMDb means in each dataframe and t-test for IMDb vs rotten tomato in streaming df  
statistical_analysis2(imdb_df2['IMDb'], streaming_df3['IMDb'])
statistical_analysis2(streaming_df3['IMDb'], streaming_df3['Rotten Tomatoes'])
```

**Data Manipulation with Pandas and Numpy**

```python
def combine_dfs(imdb, stream):
    #combine dataframes
    df = stream.append(imdb, ignore_index=True)
    df = df.set_index('Year', drop=True)
    
    #pivot table of data
    table = pd.pivot_table(df, values='IMDb', index=['Netflix', 'Hulu', 'Prime Video', 'Disney+'], columns=['Title'], aggfunc=np.mean)
    display(table)
    
    #which streaming service has the most movies per decade
    n = 0
    h = 0
    d = 0
    ap = 0
    for group, frame in df.groupby([(df.index.year//10)*10, "Netflix", "Hulu", "Prime Video", "Disney+"]):
        if group[1] == 1.0:
            n += 1
        if group[2] == 1.0:
            h += 1
        if group[3] == 1.0:
            ap += 1
        if group[4] == 1.0:
            d += 1
    print('netflix:' + str(n), 'hulu:' + str(h), 'primeVideo:' + str(ap), 'disney+:' + str(d))
    
    #which year has the highest average rating for movies
    grouped_df = df.groupby("Year")
    mean_df = grouped_df['IMDb'].mean().sort_values(ascending=False)
    print(mean_df)
    
    #which year has the highest rated movie
    g = df.groupby("Year")
    maxx = g[['IMDb', 'Title']].max().sort_values(by='IMDb', ascending=False)
    result = maxx[:5]
    lst = []
    count = 0
    for i in result.index.values:
        lst.append(i.astype(str)[0:4] + ': ' + maxx.iloc[count]['Title'])
        count += 1
    print('The years that have the highest rates movies are: ', lst, 'at a rating of ', result.iloc[0][0])
    return df

    
    
df = combine_dfs(imdb_df2, streaming_df3)
```

```python
def new_vis(df):
    #Average rating for movies each year (scatterplot)
    df = df.reset_index()
    group = df.groupby('Year').agg(np.mean)
    x = group.index.get_level_values(0)
    y = group['IMDb']
    
    plt.scatter(x, y)
    plt.xlabel("Years")
    plt.ylabel("Average IMDb Ratings")
    plt.title("Average IMDb Ratings for Each Year")
    plt.show()

    
new_vis(df)    
```

```python
def same_movie_percentage(df1,df2):
    #Makes a dataframe that has matching movie titles 
    merged_df=df1.merge(df2, how= 'inner', on = ['Title'])
    
    #Finds the total number of movies in the dataframe
    total_number_of_movies = len(merged_df.index)
    
    #Finds the number of movies for each streaming service and calcluates the total percentage
    netflix_total = merged_df['Netflix'].sum()/total_number_of_movies*100
    hulu_total = merged_df['Hulu'].sum()/total_number_of_movies*100
    prime_video_total = merged_df['Prime Video'].sum()/total_number_of_movies*100
    disney_total = merged_df['Disney+'].sum()/total_number_of_movies*100

    #creates a pie chart based off of each streaming services percentage
    pie_chart=np.array([disney_total,prime_video_total,hulu_total,netflix_total])
    pie_labels = ["Disney+","Prime Video","Hulu","Netflix"]
    plt.pie(pie_chart, labels = pie_labels, autopct='%1.1f%%')
    plt.show()
    
    
same_movie_percentage(imdb_df2, streaming_df3)
```

```python
%%sql

drop table if exists movie_avg;
create table movie_avg (
    " " varchar(255),
    "Netflix" float,
    "Hulu" float,
    "PrimeVideo" float,
    "DisneyPlus" float
);
insert into movie_avg values (
    ('average_movie_ratings:'),
    (select avg("IMDb") from streaming where "Netflix" = 1),
    (select avg("IMDb") from streaming where "Hulu" = 1),
    (select avg("IMDb") from streaming where "PrimeVideo" = 1),
    (select avg("IMDb") from streaming where "DisneyPlus" = 1)
);
select * from movie_avg;
```

```python
%%sql

select sum(combo."Netflix") as "num_netflix", sum(combo."Hulu") as "num_hulu", 
sum(combo."PrimeVideo") as "num_prime", sum(combo."DisneyPlus") as "num_disney"
from (streaming inner join imdb on streaming."Title" = imdb."Title") as combo;
```

```python
%%sql

drop table if exists platform_percentages;
create table platform_percentages (
    " " varchar(255),
    "Netflix" float,
    "Hulu" float,
    "PrimeVideo" float,
    "DisneyPlus" float
);
insert into platform_percentages values (
    ('platform_percentages_for_imdb_movies:'),
    (SELECT 
  (SELECT sum("Netflix") FROM streaming) / 
  (SELECT count("ID") FROM streaming)),
    (SELECT 
  (SELECT sum("Hulu") FROM streaming) / 
  (SELECT count("ID") FROM streaming)),
    (SELECT 
  (SELECT sum("PrimeVideo") FROM streaming) / 
  (SELECT count("ID") FROM streaming)),
    (SELECT 
  (SELECT sum("DisneyPlus") FROM streaming) / 
  (SELECT count("ID") FROM streaming))
);

select * from platform_percentages;
```
